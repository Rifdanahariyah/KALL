
## **Matriks Invers**
---

## **Definisi Matriks Invers**
Matriks invers adalah matriks yang ketika dikalikan dengan matriks asalnya menghasilkan matriks identitas. 

Matriks invers digunakan untuk menyelesaikan sistem persamaan linear, transformasi geometri dan tidak semua matriks memiliki invers, hanya matriks dengan determinannya tidak nol yang bisa dihitung inversnya.

---
## **Tugas mencari matriks invers**
##### **1. Menyusun Sistem Persamaan dalam Bentuk Matriks**
$$
\begin{aligned}  
7x_1 - 4x_2 + 3x_3 &= 5 \\  
2x_1 + 5x_2 - 6x_3 &= 7 \\  
-3x_1 + 8x_2 + x_3 &= 3  
\end{aligned}
$$

Sistem persamaan ini dapat ditulis dalam bentuk matriks sebagai:

$$
A x = b
$$

dengan:

$$
A = 
\begin{bmatrix}  
7 & -4 & 3 \\  
2 & 5 & -6 \\  
-3 & 8 & 1  
\end{bmatrix}
$$

adalah matriks koefisien,

$$
x = 
\begin{bmatrix}  
x_1 \\  
x_2 \\  
x_3  
\end{bmatrix}
$$

adalah vektor variabel, dan

$$
b = 
\begin{bmatrix}  
5 \\  
7 \\  
3  
\end{bmatrix}
$$

adalah vektor hasil.

##### **2. Mencari Matriks Invers $ A^-1 $**
Untuk menyelesaikan sistem ini menggunakan metode matriks invers, kita harus mencari $ A^{-1} $ sehingga:

$$
x = A^{-1} b
$$

Misalkan hasil perhitungan invers dari \( A \) adalah:

$$
A^{-1} =
\begin{bmatrix}  
\frac{7}{15} & \frac{-4}{15} & \frac{3}{15} \\  
\frac{2}{15} & \frac{5}{15} & \frac{-6}{15} \\  
\frac{-3}{15} & \frac{8}{15} & \frac{1}{15}  
\end{bmatrix}
$$

##### **3. Mengalikan Matriks Invers**
Sekarang kita kalikan \( A^{-1} \) dengan \( b \):

$$
x = A^{-1} b
$$

$$
\begin{bmatrix}  
\frac{7}{15} & \frac{-4}{15} & \frac{3}{15} \\  
\frac{2}{15} & \frac{5}{15} & \frac{-6}{15} \\  
\frac{-3}{15} & \frac{8}{15} & \frac{1}{15}  
\end{bmatrix}
\begin{bmatrix}  
5 \\  
7 \\  
3  
\end{bmatrix}
$$

Lakukan perkalian baris dengan kolom:

$$
x_1 = \left( \frac{7}{15} \times 5 \right) + \left( \frac{-4}{15} \times 7 \right) + \left( \frac{3}{15} \times 3 \right)
$$

$$
= \frac{35}{15} - \frac{28}{15} + \frac{9}{15} = \frac{16}{15} \approx 1.067
$$

$$
x_2 = \left( \frac{2}{15} \times 5 \right) + \left( \frac{5}{15} \times 7 \right) + \left( \frac{-6}{15} \times 3 \right)
$$

$$
= \frac{10}{15} + \frac{35}{15} - \frac{18}{15} = \frac{27}{15} \approx 0.067
$$

$$
x_3 = \left( \frac{-3}{15} \times 5 \right) + \left( \frac{8}{15} \times 7 \right) + \left( \frac{1}{15} \times 3 \right)
$$

$$
= \frac{-15}{15} + \frac{56}{15} + \frac{3}{15} = \frac{44}{15} \approx 0.467
$$

Sehingga solusi sistem adalah:

$$
x =
\begin{bmatrix}  
\frac{16}{15} \\  
\frac{27}{15} \\  
\frac{44}{15}  
\end{bmatrix}
\approx
\begin{bmatrix}  
1.067 \\  
0.067 \\  
0.467  
\end{bmatrix}
$$

##### **4. Hasil Akhir**

Hasil akhir dari perhitungan adalah:

$$
x =
\begin{bmatrix}  
1.067 \\  
0.067 \\  
0.467  
\end{bmatrix}
$$

Jika dibulatkan ke bilangan bulat terdekat:

$$
x =
\begin{bmatrix}  
1 \\  
0 \\  
0  
\end{bmatrix}
$$

Sehingga solusi sistem adalah : 
$$
x_1 = 1, \quad x_2 = 0, \quad x_3 = 0
$$

\begin{bmatrix} 5 \\ 7 \\ 3 \end{bmatrix}=
\begin{bmatrix} 1.067 \\ 0.067 \\ 0.467 \end{bmatrix}=
\begin{bmatrix} 1 \\ 0 \\ 0 \end{bmatrix}$$
